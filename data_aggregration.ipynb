```python
def join_with_default(master_df, other_df, on_column, new_columns, default_value):
    
    columns_to_merge = [on_column] + new_columns
    result_df = pd.merge(master_df, other_df[columns_to_merge], 
                         on=on_column, how='left')
    
    # Fill NaN values in the new columns with the default value
    for col in new_columns:
        result_df[col] = result_df[col].fillna(default_value)
    
    return result_df
```

```python
result = join_with_default(new_df, ondf, 'DATE', 'value', 0)
```
