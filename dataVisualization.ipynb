```python

def convert_to_number(s):
    return pd.to_numeric(s.str.replace('"', '').str.replace(',', ''), errors='coerce')

# Apply the conversion to your column
df['shares'] = convert_to_number(df['shares'])
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Ensure 'DATE' is in datetime format
# df['DATE'] = pd.to_datetime(df['DATE'])

# List of columns to plot
other_columns = ['gdp_growth', 'house_inventory', 'Diff_house_inventory', 
                 'inflation', 'material_cost', 'mortgage_rate', 'population_growth', 
                 'shares', 'Unemployment', 'vacancy_rate', 'Total', '1 unit', 
                 '2 to 4', '5 units', '5 ratio', 'house_price', 'house_price_diff']

# Normalize the data
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[other_columns]), columns=other_columns)
df_normalized['DATE'] = df['DATE']

num_plots = len(other_columns) - 2  # excluding 'house_price' and 'house_price_diff'
fig, axes = plt.subplots(num_plots, 1, figsize=(12, 3*num_plots), sharex=True)

# Plotting each normalized variable along with normalized house_price and house_price_diff
for i, column in enumerate(other_columns[:-2]):  # Exclude 'house_price' and 'house_price_diff' from loop
    ax = axes[i]
    ax.plot(df_normalized['DATE'], df_normalized[column], label=column)
    ax.plot(df_normalized['DATE'], df_normalized['house_price'], label='house_price', linestyle='--', alpha=0.7)
    ax.plot(df_normalized['DATE'], df_normalized['house_price_diff'], label='house_price_diff', linestyle='-.', alpha=0.7)
    ax.set_ylabel(column)
    ax.legend(loc='upper right')
    ax.grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(num_plots, 1, figsize=(12, 3*num_plots), sharex=True)

# Plotting each variable with a secondary y-axis for house_price and house_price_diff
for i, column in enumerate(other_columns[:-2]):  # Exclude 'house_price' and 'house_price_diff' from loop
    ax = axes[i]
    ax.plot(df['DATE'], df[column], label=column)
    ax.set_ylabel(column)
    ax.grid(True)
    ax2 = ax.twinx()
    ax2.plot(df['DATE'], df['house_price'], label='house_price', linestyle='--', alpha=0.7, color='orange')
    ax2.plot(df['DATE'], df['house_price_diff'], label='house_price_diff', linestyle='-.', alpha=0.7, color='green')
    ax2.set_ylabel('House Price and Diff')
    ax.legend(loc='upper left')
    ax2.legend(loc='upper right')

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
